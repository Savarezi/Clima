## Ajustar Tipos de Dados e Fuso Horário


* Converte a coluna 'time' para o tipo datetime, garantindo que as datas sejam corretamente interpretadas no fuso horário 'America/Sao_Paulo', conforme especificado na URL dos dados. Confirma a aplicação do fuso horário na coluna de tempo.


```python
df['time'] = pd.to_datetime(df['time'])
df['time'] = df['time'].dt.tz_localize('America/Sao_Paulo')
df.info()
```

## Traduzir e Renomear Colunas


Traduz os nomes das colunas principais (e.g., 'time', 'temperature_2m_max (°C)', 'temperature_2m_min (°C)', 'precipitation_sum (mm)') para português, para facilitar a compreensão e apresentação dos resultados.


```python
column_mapping = {
    'time': 'Data',
    'temperature_2m_max (°C)': 'Temperatura Máxima (°C)',
    'temperature_2m_min (°C)': 'Temperatura Mínima (°C)',
    'precipitation_sum (mm)': 'Precipitação Total (mm)'
}
df.rename(columns=column_mapping, inplace=True)
print(df.columns)
```

## Visualização Interativa dos Dados de Tempo


* Gera um gráfico de linha interativo que mostra a evolução diária da temperatura máxima, temperatura mínima e a soma de precipitação ao longo do período, com legendas claras e títulos em português, para uma compreensão visual das tendências.

* Para criar um gráfico interativo que visualize a evolução diária da temperatura máxima, temperatura mínima e precipitação total, utilizarei a biblioteca `plotly.graph_objects` e `plotly.subplots.make_subplots`. Esta abordagem permite ter dois eixos Y independentes (um para temperatura e outro para precipitação) em um único gráfico interativo, melhorando a análise visual. Adicionarei legendas, títulos e rótulos de eixo em português para garantir clareza.

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Cria uma figura com dois eixos Y
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Adiciona o traço para Temperatura Máxima
fig.add_trace(
    go.Scatter(x=df['Data'], y=df['Temperatura Máxima (°C)'], name='Temperatura Máxima (°C)', mode='lines', line=dict(color='red')),
    secondary_y=False,
)

# Adiciona o traço para Temperatura Mínima
fig.add_trace(
    go.Scatter(x=df['Data'], y=df['Temperatura Mínima (°C)'], name='Temperatura Mínima (°C)', mode='lines', line=dict(color='blue')),
    secondary_y=False,
)

# Adiciona o traço para Precipitação Total
fig.add_trace(
    go.Scatter(x=df['Data'], y=df['Precipitação Total (mm)'], name='Precipitação Total (mm)', mode='lines', line=dict(color='green', dash='dot')),
    secondary_y=True,
)

# Configura o layout do gráfico
fig.update_layout(
    title_text='<b>Evolução Diária de Temperatura e Precipitação</b>',
    # Define o título do gráfico em negrito
    hovermode='x unified'  # Mostra informações de todas as séries no hover
)

# Configura o eixo Y primário (Temperatura)
fig.update_yaxes(
    title_text='<b>Temperatura (°C)</b>',  # Rótulo do eixo Y primário em negrito
    secondary_y=False,
)

# Configura o eixo Y secundário (Precipitação)
fig.update_yaxes(
    title_text='<b>Precipitação (mm)</b>',  # Rótulo do eixo Y secundário em negrito
    secondary_y=True,
)

# Configura o eixo X (Data)
fig.update_xaxes(
    title_text='<b>Data</b>'  # Rótulo do eixo X em negrito
)

fig.show()
```
